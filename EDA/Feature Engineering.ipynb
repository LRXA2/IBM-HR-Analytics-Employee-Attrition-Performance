# EDA for continous features

```python
ordinal_cols = ['Education', 'JobLevel', 'StockOptionLevel', 'RelationshipSatisfaction', 'TrainingTimesLastYear', 'NumCompaniesWorked']

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
continuous_cols = [col for col in numeric_cols if col not in ordinal_cols]

categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols += ordinal_cols

for col in continuous_cols:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # --- Left: Box plot ---
    sns.boxplot(x='Attrition', y=col, data=df, ax=axes[0])
    axes[0].set_title(f'{col} vs Attrition')
    axes[0].set_xlabel('Attrition')
    axes[0].set_ylabel(col)

    # --- Right: Distribution plot ---
    sns.histplot(df[df['Attrition'] == 'Yes'][col], label='Attrition: Yes', kde=True, stat='density', ax=axes[1])
    sns.histplot(df[df['Attrition'] == 'No'][col], label='Attrition: No', kde=True, stat='density', ax=axes[1])
    axes[1].set_title(f'{col} Distribution by Attrition')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Density')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

results = []

for col in continuous_cols:
    # Skip constant columns
    if df[col].nunique() <= 1:
        continue

    # Split groups
    yes = df[df['Attrition'] == 'Yes'][col].dropna()
    no  = df[df['Attrition'] == 'No'][col].dropna()

    # Safety check
    if len(yes) < 2 or len(no) < 2:
        continue

    # Welch's t-test
    t_stat, p_t = ttest_ind(yes, no, equal_var=False)

    # Mannâ€“Whitney U
    u_stat, p_u = mannwhitneyu(yes, no, alternative='two-sided')

    # Cohen's d
    mean_diff = yes.mean() - no.mean()
    pooled_std = np.sqrt((yes.var() + no.var()) / 2)
    cohens_d = mean_diff / pooled_std if pooled_std != 0 else np.nan

    results.append({
        'Feature': col,
        'Welch_t_p': p_t,
        'MannWhitney_p': p_u,
        'Cohens_d': cohens_d,
        'Significant (t)': p_t < 0.05,
        'Significant (MW)': p_u < 0.05
    })

# Results dataframe
results_df = pd.DataFrame(results)

# Sort by effect size magnitude
results_df['|Cohens_d|'] = results_df['Cohens_d'].abs()
results_df = results_df.sort_values('|Cohens_d|', ascending=False)

results_df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
import pandas as pd
import numpy as np

models = {
    'Age': ['Age'],
    'TotalWorkingYears': ['TotalWorkingYears'],
    'MonthlyIncome': ['MonthlyIncome'],
    'Age + MonthlyIncome': ['Age', 'MonthlyIncome'],
    'TWY + MonthlyIncome': ['TotalWorkingYears', 'MonthlyIncome'],
    'Age + TWY': ['Age', 'TotalWorkingYears'],
    'All three': ['Age', 'TotalWorkingYears', 'MonthlyIncome']
}

X = df[['Age', 'TotalWorkingYears', 'MonthlyIncome']]
y = df['Attrition'].map({'Yes': 1, 'No': 0})

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = []

for name, cols in models.items():
    model = LogisticRegression(max_iter=1000, solver='liblinear')
    auc = cross_val_score(
        model, X[cols], y,
        cv=cv,
        scoring='roc_auc'
    ).mean()
    results.append((name, auc))

results_df = pd.DataFrame(results, columns=['Model', 'Mean ROC-AUC'])
results_df.sort_values('Mean ROC-AUC', ascending=False)
```

## Justification for Grouping Ordinal Features

Several ordinal features in the dataset exhibit **non-linear and non-monotonic relationships** with employee attrition. While these variables have an inherent ordering, treating them as purely numeric in a logistic regression model implicitly assumes a **linear and monotonic effect**, which was not supported by exploratory data analysis.

### Motivation for Grouping

Attrition rate analysis by category revealed that:
- Some ordinal features show **threshold effects** (e.g., RelationshipSatisfaction).
- Others exhibit **non-linear or clustered risk patterns** (e.g., StockOptionLevel, TrainingTimesLastYear, NumCompaniesWorked).
- Numeric ordinal encoding in these cases risks **misrepresenting the true relationship** between the feature and attrition by forcing a linear trend.

To address this, categories with **similar attrition behavior** were grouped together. This approach:
- Removes false linear assumptions
- Reduces noise caused by sparse or unstable categories
- Improves interpretability without sacrificing predictive performance

---

### Grouping vs. Data Leakage

Grouping decisions were informed by **exploratory analysis only** and were applied at a **coarse level**, without encoding exact attrition probabilities or rankings. This differs fundamentally from target-based re-ranking or target encoding and therefore **does not constitute data leakage**.

Importantly:
- Grouping preserves the original semantic meaning of the features where applicable
- No information unavailable at prediction time was used
- Group definitions were fixed prior to model evaluation

---

### Encoding Strategy After Grouping

After grouping:
- Features that retained a **true semantic order** (e.g., Education, JobLevel) were encoded ordinally
- Features whose groupings reflected **risk-based or threshold effects** were treated as categorical (binary or one-hot encoded)

This hybrid encoding strategy ensures that each feature is represented in a manner consistent with its underlying structure and relationship to attrition.

---

### Impact on Model Performance

Grouped and original encodings were compared using:
- ROC-AUC
- Permutation importance
- SHAP values

Results showed that grouped features:
- Preserved or improved predictive performance
- Produced more stable and interpretable feature contributions
- Reduced contradictory or noisy effects observed in the original ordinal encoding

```python
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# --------------------------
# 0) Target
# --------------------------
y = df['Attrition'].map({'Yes': 1, 'No': 0})

# --------------------------
# 1) Continuous baseline features (controls)
# --------------------------
cont_cols = ['Age', 'DistanceFromHome']

missing_cont = [c for c in cont_cols if c not in df.columns]
if missing_cont:
    raise KeyError(f"Missing continuous columns: {missing_cont}")

X_base = df[cont_cols].copy()

# --------------------------
# 2) Model A: ORIGINAL ordinal encoding
# --------------------------
ord_original_cols = [
    'BusinessTravel_Ordinal',
    'JobLevel',
    'Education',
    'StockOptionLevel',
    'RelationshipSatisfaction',
    'TrainingTimesLastYear',
    'NumCompaniesWorked'
]

missing_ord_orig = [c for c in ord_original_cols if c not in df.columns]
if missing_ord_orig:
    raise KeyError(f"Missing original ordinal columns: {missing_ord_orig}")

X_original = pd.concat([
    X_base,
    df[ord_original_cols]
], axis=1)

# --------------------------
# 3) Model B: GROUPED ordinal encoding
# --------------------------
required_grouped = [
    'BusinessTravel_Ordinal',          # unchanged
    'JobLevel_grouped_ord',            # ordinal
    'Education_grouped_ord',           # ordinal
    'RelationshipSatisfaction_bin',    # binary
    'StockOption_grouped_cat',         # categorical -> one-hot
    'TrainingTimesLastYear_grouped_cat',# categorical -> one-hot
    'NumCompaniesWorked_grouped_cat'   # categorical -> one-hot
]

missing_grp = [c for c in required_grouped if c not in df.columns]
if missing_grp:
    raise KeyError(f"Missing grouped columns: {missing_grp}. Create them first, then rerun.")

X_grouped = pd.concat([
    X_base,
    df[['BusinessTravel_Ordinal',
        'JobLevel_grouped_ord',
        'Education_grouped_ord',
        'RelationshipSatisfaction_bin']],
    pd.get_dummies(df[['StockOption_grouped_cat',
                       'TrainingTimesLastYear_grouped_cat',
                       'NumCompaniesWorked_grouped_cat']], drop_first=False)
], axis=1)

# --------------------------
# 4) Scale continuous columns (outside pipeline)
# --------------------------
scaler = StandardScaler()

X_original_scaled = X_original.copy()
X_grouped_scaled  = X_grouped.copy()

# Fit on original, apply to both (consistent scaling for A/B comparison)
X_original_scaled[cont_cols] = scaler.fit_transform(X_original_scaled[cont_cols])
X_grouped_scaled[cont_cols]  = scaler.transform(X_grouped_scaled[cont_cols])

# Optional sanity checks
print("\nScaled summary (original):")
print(X_original_scaled[cont_cols].describe().loc[['mean','std']])

print("\nScaled summary (grouped):")
print(X_grouped_scaled[cont_cols].describe().loc[['mean','std']])

print("X_original shape:", X_original.shape)
print("X_grouped shape :", X_grouped.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import permutation_importance
from sklearn.metrics import roc_auc_score
import shap

model_original = LogisticRegression(max_iter=1000, random_state=42)
model_grouped  = LogisticRegression(max_iter=1000, random_state=42)

model_original.fit(X_original_scaled, y)
model_grouped.fit(X_grouped_scaled, y)
```

```python
roc_original = roc_auc_score(y, model_original.predict_proba(X_original)[:,1])
roc_grouped = roc_auc_score(y, model_grouped.predict_proba(X_grouped)[:,1])

roc_original, roc_grouped
```

```python
perm_original = permutation_importance(
    model_original, X_original, y,
    scoring='roc_auc', n_repeats=10, random_state=42
)

perm_grouped = permutation_importance(
    model_grouped, X_grouped, y,
    scoring='roc_auc', n_repeats=10, random_state=42
)

perm_original_df = pd.Series(
    perm_original.importances_mean,
    index=X_original.columns
).sort_values(ascending=False)

perm_grouped_df = pd.Series(
    perm_grouped.importances_mean,
    index=X_grouped.columns
).sort_values(ascending=False)

perm_original_df, perm_grouped_df
```

## Feature Engineering & Grouping Summary

**Objective**
Improve model performance, stability, and interpretability by correcting mis-specified ordinal features, reducing noise, and grouping categories based on observed attrition behavior rather than raw numeric order.

---

### Method

Each feature was evaluated using:

* ROC AUC comparison
* Permutation importance
* SHAP values

Decisions were only made when conclusions were consistent across methods.

---

### Feature Decisions

* **Education**
  Dropped. Shows near-zero permutation importance and SHAP contribution in all encodings, indicating no incremental predictive value after controlling for other features.

* **JobLevel**
  Grouped and kept as an ordinal feature. Grouping by organizational hierarchy (entry / mid / senior) significantly increased importance and produced stable, monotonic effects.

* **StockOptionLevel**
  Grouped into `0` vs `1â€“2`, with level `3` treated as an implicit reference category. The highest level is rare and shows negligible independent contribution; grouping consolidates signal and improves stability.

* **RelationshipSatisfaction**
  Retained as a binary feature (`1` vs `2â€“4`). Shows a weak but consistent effect and is more stable than ordinal encoding.

* **TrainingTimesLastYear**
  Regrouped based on observed attrition rates rather than numeric order. Attrition risk is non-monotonic, with high risk at `{0, 4}` and lower risk at `{1, 3, 5, 6}`.
  The intermediate group (`TT_mid`, value `2`) shows near-zero permutation importance and SHAP contribution and is therefore dropped to reduce noise and simplify the model.

* **NumCompaniesWorked**
  Retained as grouped categorical feature. Exhibits non-linear behavior; categorical grouping outperforms ordinal encoding and aligns with career mobility stages.

* **BusinessTravel**
  Kept unchanged. Shows consistent, interpretable contribution across all models.

```python
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape

    # Bias correction (recommended)
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2) / (n - 1)
    kcorr = k - ((k - 1)**2) / (n - 1)

    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))
```

```python
results = []

for col in nominal_cols:
    contingency = pd.crosstab(df[col], df['Attrition'])
    v = cramers_v(contingency)

    results.append({
        'Feature': col,
        'Cramers_V': v
    })

pd.DataFrame(results).sort_values('Cramers_V', ascending=False)
```

## Nominal Feature Encoding & Simplification â€“ Summary

Several nominal variables in the dataset (e.g. *JobRole*, *EducationField*, *MaritalStatus*, *Department*) originally had multiple categories. Direct one-hot encoding increased dimensionality and diluted predictive signal. To address this, categories were **grouped and simplified based on empirical evidence**, with the goal of preserving predictive power while improving interpretability.

Encoding choices were evaluated using **logistic regression**, **permutation importance (ROC AUC)**, and **SHAP values**.

### Key decisions

* **OverTime** was encoded as binary (*Yes = 1, No = 0*) and consistently emerged as the strongest predictor across all models.
* **MaritalStatus** was simplified to *Single vs Not Single*. This binary encoding increased both permutation importance and SHAP impact, indicating more concentrated signal.
* **Department** was encoded as *Sales vs Non-Sales*. Fine-grained department categories showed minimal additional value.
* **EducationField** was grouped into *High / Medium / Low*, then simplified to *Low vs Other* after *Medium* showed negligible importance.
* **JobRole** was grouped into attrition-risk buckets and ultimately simplified into **High / Medium / Low** risk categories. This retained most of the original predictive signal while substantially reducing complexity.

```python
cont_cols = ['MonthlyIncome', 'Age', 'DistanceFromHome', 'DailyRate']

oridnal_cols = ['BusinessTravel', 'JobLevel', 'StockOptionLevel', 'RelationshipSatisfaction', 'TrainingTimesLastYear', 'NumCompaniesWorked']

nominal_cols = ['Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']

eda_df = pd.read_csv(DATA_PATH+'main_dataset.csv')
eda_df = eda_df[cont_cols + oridnal_cols + nominal_cols]

business_travel_map = {
    'Non-Travel': 0,
    'Travel_Rarely': 1,
    'Travel_Frequently': 2}

job_level_map = {
    1: 0,
    3: 1,
    2: 2, 4: 2, 5: 2}

sol_map = {0: 'SOL_0',
    1: 'SOL_12', 2: 'SOL_12'}

rs_map = {
    1: 1,
    2: 1, 3: 1, 4: 1}

tt_map = {
    0: 'high', 4: 'high',
    3: 'low', 1: 'low', 5: 'low', 6: 'low'}

ncw_map = {
    1: 'NCW_1', 
    5: 'NCW_5679', 6: 'NCW_5679', 7: 'NCW_5679', 9: 'NCW_5679',
    8: 'NCW_84023', 4: 'NCW_84023', 0: 'NCW_84023', 2: 'NCW_84023', 3: 'NCW_84023'}

department_map = {
    'Sales': 1,
    'Human Resources': 0,
    'Research & Development': 0    
}

ef_map = {
    'Human Resources': 'pos',
    'Technical Degree': 'pos',
    'Marketing': 'pos',
    'Life Sciences': 'neg',
    'Medical': 'neg',
    'Other': 'neg'    
}

job_role_map = {
    'Sales Representative': 'High',
    'Laboratory Technician': 'High',
    'Human Resources': 'High',
    'Sales Executive': 'Medium',
    'Research Scientist': 'Low',
    'Manufacturing Director': 'Low',
    'Healthcare Representative': 'Low',
    'Manager': 'Low',
    'Research Director': 'Low'
}

ms_map = {
    'Single': 1,
    'Married': 0,
    'Divorced': 0    
}

eda_df['BusinessTravel_ord'] = eda_df['BusinessTravel'].map(business_travel_map)
eda_df['JobLevel_ord'] = eda_df['JobLevel'].map(job_level_map)
eda_df['RelationshipSatisfaction_bin'] = eda_df['RelationshipSatisfaction'].map(rs_map)
eda_df['TrainingTimesLastYear_grp'] = eda_df['TrainingTimesLastYear'].map(tt_map)
eda_df['NumCompaniesWorked_grp'] = eda_df['NumCompaniesWorked'].map(ncw_map)
eda_df['Department_binary'] = eda_df['Department'].map(department_map)
eda_df['EducationField_grp'] = eda_df['EducationField'].map(ef_map)
eda_df['JobRole_grp'] = eda_df['JobRole'].map(job_role_map)
eda_df['MaritalStatus_binary'] = eda_df['MaritalStatus'].map(ms_map)
eda_df['OverTime_numeric'] = eda_df['OverTime'].map({'Yes': 1, 'No': 0})

eda_df.to_csv(DATA_PATH+'eda_data.csv', index=False)
```

# EDA for continous features

```python
ordinal_cols = ['Education', 'JobLevel', 'StockOptionLevel', 'RelationshipSatisfaction', 'TrainingTimesLastYear', 'NumCompaniesWorked']

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
continuous_cols = [col for col in numeric_cols if col not in ordinal_cols]

categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols += ordinal_cols

for col in continuous_cols:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # --- Left: Box plot ---
    sns.boxplot(x='Attrition', y=col, data=df, ax=axes[0])
    axes[0].set_title(f'{col} vs Attrition')
    axes[0].set_xlabel('Attrition')
    axes[0].set_ylabel(col)

    # --- Right: Distribution plot ---
    sns.histplot(df[df['Attrition'] == 'Yes'][col], label='Attrition: Yes', kde=True, stat='density', ax=axes[1])
    sns.histplot(df[df['Attrition'] == 'No'][col], label='Attrition: No', kde=True, stat='density', ax=axes[1])
    axes[1].set_title(f'{col} Distribution by Attrition')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Density')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

results = []

for col in continuous_cols:
    # Skip constant columns
    if df[col].nunique() <= 1:
        continue

    # Split groups
    yes = df[df['Attrition'] == 'Yes'][col].dropna()
    no  = df[df['Attrition'] == 'No'][col].dropna()

    # Safety check
    if len(yes) < 2 or len(no) < 2:
        continue

    # Welch's t-test
    t_stat, p_t = ttest_ind(yes, no, equal_var=False)

    # Mannâ€“Whitney U
    u_stat, p_u = mannwhitneyu(yes, no, alternative='two-sided')

    # Cohen's d
    mean_diff = yes.mean() - no.mean()
    pooled_std = np.sqrt((yes.var() + no.var()) / 2)
    cohens_d = mean_diff / pooled_std if pooled_std != 0 else np.nan

    results.append({
        'Feature': col,
        'Welch_t_p': p_t,
        'MannWhitney_p': p_u,
        'Cohens_d': cohens_d,
        'Significant (t)': p_t < 0.05,
        'Significant (MW)': p_u < 0.05
    })

# Results dataframe
results_df = pd.DataFrame(results)

# Sort by effect size magnitude
results_df['|Cohens_d|'] = results_df['Cohens_d'].abs()
results_df = results_df.sort_values('|Cohens_d|', ascending=False)

results_df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
import pandas as pd
import numpy as np

models = {
    'Age': ['Age'],
    'TotalWorkingYears': ['TotalWorkingYears'],
    'MonthlyIncome': ['MonthlyIncome'],
    'Age + MonthlyIncome': ['Age', 'MonthlyIncome'],
    'TWY + MonthlyIncome': ['TotalWorkingYears', 'MonthlyIncome'],
    'Age + TWY': ['Age', 'TotalWorkingYears'],
    'All three': ['Age', 'TotalWorkingYears', 'MonthlyIncome']
}

X = df[['Age', 'TotalWorkingYears', 'MonthlyIncome']]
y = df['Attrition'].map({'Yes': 1, 'No': 0})

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = []

for name, cols in models.items():
    model = LogisticRegression(max_iter=1000, solver='liblinear')
    auc = cross_val_score(
        model, X[cols], y,
        cv=cv,
        scoring='roc_auc'
    ).mean()
    results.append((name, auc))

results_df = pd.DataFrame(results, columns=['Model', 'Mean ROC-AUC'])
results_df.sort_values('Mean ROC-AUC', ascending=False)
```

**Outcome:**
- Model is concise, interpretable, and retains all meaningful predictors of attrition.

```python
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape)
    return np.sqrt(chi2 / (n * (k - 1)))

results = []

for col in ordinal_cols:
    contingency = pd.crosstab(df[col], df['Attrition'])
    v = cramers_v(contingency)

    results.append({
        'Feature': col,
        'Cramers_V': v
    })

pd.DataFrame(results).sort_values('Cramers_V', ascending=False)
```

```python
from sklearn.inspection import permutation_importance

X_ord = df[ordinal_cols]
y = df['Attrition'].map({'Yes': 1, 'No': 0})

model = LogisticRegression(max_iter=1000)
model.fit(X_ord, y)

perm = permutation_importance(
    model,
    X_ord,
    y,
    scoring='roc_auc',
    n_repeats=10,
    random_state=42
)

pd.Series(
    perm.importances_mean,
    index=X_ord.columns
).sort_values(ascending=False)
```

### Ordinal Feature Analysis and Feature Selection

To evaluate the relevance of ordinal features for predicting employee attrition, two tests were conducted: **CramÃ©râ€™s V** and **Permutation Importance**.

1. **CramÃ©râ€™s V**  
   - Measures the strength of association between categorical/ordinal features and the target variable (Attrition).  
   - Values range from 0 (no association) to 1 (perfect association).  
   - Results indicated that features such as `JobLevel` and `StockOptionLevel` had moderate associations, whereas `Education` (0.046) and `RelationshipSatisfaction` (0.060) showed very weak associations, suggesting minimal predictive signal on their own.

2. **Permutation Importance**  
   - Evaluates the contribution of each feature to model performance by measuring the decrease in ROC-AUC when the feature is randomly shuffled.  
   - In the logistic regression model, `Education` had almost zero importance (0.0003) and `RelationshipSatisfaction` had very low importance (0.0074), confirming that the model does not rely on these features for prediction.

**Justification for Dropping Features**  
Based on the consistent results from both tests, `Education` and `RelationshipSatisfaction` were removed from the ordinal feature set. Dropping these features reduces dataset complexity, avoids potential noise, and focuses the analysis on more meaningful predictors without losing predictive power.

### Likelihood Ratio Test and Ordinal Feature Analysis

To evaluate the incremental value of ordinal features in predicting employee attrition, we conducted **Likelihood Ratio Tests (LRTs)** using logistic regression.

1. **Adding `NumCompaniesWorked`**
   - We compared a base model with `JobLevel`, `StockOptionLevel`, and `BusinessTravel_Ordinal` against a model that also included `NumCompaniesWorked`.  
   - **Result:** p-value = 0.00175 (< 0.01), indicating that adding `NumCompaniesWorked` significantly improves the model.  
   - **Interpretation:** This feature provides meaningful incremental information about attrition and should be retained.

2. **Adding `TrainingTimesLastYear`**
   - Using the same base model, we tested the effect of including `TrainingTimesLastYear`.  
   - **Result:** p-value = 0.00952 (< 0.01), which is statistically significant but borderline under a strict 0.01 threshold.  
   - **Coefficient:** -0.151 â†’ the odds ratio `exp(-0.151) â‰ˆ 0.86` indicates that each additional training session **reduces the odds of attrition by ~14%**.  
   - **Interpretation:** The feature has a small but real effect. Its contribution is modest, but it is not redundant with other ordinal features.

3. **Correlation Analysis**
   - A correlation heatmap of all ordinal features showed a **maximum correlation of 0.14**, indicating **very low multicollinearity**.  
   - **Interpretation:** Including `NumCompaniesWorked` or `TrainingTimesLastYear` in the model does not introduce redundancy or instability.

### Justification for Feature Retention
- **NumCompaniesWorked:** Strong statistical evidence and meaningful effect â†’ **keep**.  
- **TrainingTimesLastYear:** Small effect but statistically significant and not redundant â†’ **keep for completeness**, especially if the goal is to maximize predictive coverage.  
- **Correlation check** confirms that adding these features does not introduce multicollinearity, supporting their inclusion.
